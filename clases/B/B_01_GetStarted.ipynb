## Creacion del entorno FMP

Proveemos un archivo de entorno el cual presenta una lista con todos los paquetes requeridos para interactuar con los cuadernos FMP. Abra una ventana de comandos en Windows o terminal en Linux/macOS. Desde el directorio que contiene el archivo de entorno `environment.yml` ejecute lo siguiente: 
<div class="alert alert-block alert-warning">
<tt>conda env create -f environment.yml</tt>
</div> 

## Lanzamiento del servidor Jupyter 

A continuacion, activamos el entorno (*ambiente*) FMP con el siguiente comando:

<div class="alert alert-block alert-warning">
<tt>conda activate FMP</tt>
</div> 

Finalmente, desde el directorio que contiene los cuadernos FMP
ejecute lo siguiente para abrir una sesi칩n de Jupyter en su navegador web:

<div class="alert alert-block alert-warning">
<tt>jupyter notebook</tt>
</div> 

Este comando permitira mostrar la estructura de los cuadernos FMP. Asi, sera posible abrir cualquier cuaderno FMP seleccionando un archivo con la extension `.ipynb`.
